```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    'Category': np.repeat(['Math', 'Science', 'English'], 30),
    'Score': np.concatenate([
        np.random.normal(75, 10, 30),
        np.random.normal(80, 8, 30),
        np.random.normal(70, 12, 30)
    ])
}
df = pd.DataFrame(data)

sns.set_theme(style='darkgrid', palette='pastel')

cat_plot = sns.catplot(
    data=df,
    x='Category',
    y='Score',
    kind='box',
    palette='Set2',
    height=6, aspect=1.2
)

cat_plot.fig.suptitle('Exam Results by Category', fontsize=18, fontweight='bold')
cat_plot.set_axis_labels('Subject Category', 'Exam Score', fontsize=14)
cat_plot.despine(left=True)

for ax in cat_plot.axes.flat:
    ax.grid(True, linestyle='--', linewidth=0.7, color='gray', alpha=0.7)
    ax.tick_params(axis='both', labelsize=12)

plt.subplots_adjust(top=0.90)

plot_filename = 'custom_catplot.png'
plt.savefig(plot_filename)

plot_filename
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'RM': [6.5, 7.2, 6.1, 5.6, 7.0, 6.7, 6.0, 5.9, 6.8, 5.2],
    'LSTAT': [4.98, 9.14, 4.03, 2.94, 5.33, 5.21, 12.43, 15.34, 10.64, 8.78],
    'PTRATIO': [15.3, 17.8, 18.7, 19.2, 17.8, 18.7, 19.2, 15.3, 17.8, 18.7],
    'MEDV': [24, 21.6, 34.7, 33.4, 36.2, 28.7, 22.9, 27.1, 16.5, 18.9],
    'AGE': [65.2, 78.9, 61.1, 45.8, 58.7, 67.2, 84.5, 31.7, 41.3, 48.5]
}
df = pd.DataFrame(data)
corr = df.corr()
plt.figure(figsize=(7, 5))
sns.set_theme(style='white')
ax = sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.7, linecolor='gray', cbar_kws={'shrink': .9})
plt.title('Correlation Matrix of Housing Features', fontsize=16, fontweight='bold')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12, rotation=0)
plt.tight_layout()
plt.savefig('housing_corr_heatmap.png')
```
