```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create Department Budget DataFrame
budget_data = {
    'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Sales'],
    'Budget': [50000, 120000, 80000, 60000, 100000]
}
df_budget = pd.DataFrame(budget_data)

# Step 2: Create Actual Expenses DataFrame
expenses_data = {
    'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Sales'],
    'Actual': [52000, 110000, 85000, 63000, 98000]
}
df_expenses = pd.DataFrame(expenses_data)

# Step 3: Merge DataFrames on 'Department'
df_merged = pd.merge(df_budget, df_expenses, on='Department')

# Step 4: Calculate Variance (Actual - Budget)
df_merged['Variance'] = df_merged['Actual'] - df_merged['Budget']

# Step 5: Identify Departments Exceeding Budget
df_merged['Over Budget'] = df_merged['Variance'] > 0

# Step 6: Sort by Variance (Descending)
df_sorted = df_merged.sort_values(by='Variance', ascending=False)

# Step 7 (Optional): Plot Budget vs Actual using bar chart
def plot_budget_vs_actual(df):
    plt.figure(figsize=(10, 6))
    x = df['Department']
    bar1 = plt.bar(x, df['Budget'], label='Budget')
    bar2 = plt.bar(x, df['Actual'], label='Actual', alpha=0.7)

    plt.title('Department Budget vs Actual Expenses')
    plt.xlabel('Department')
    plt.ylabel('Amount')
    plt.legend()
    plt.tight_layout()
    plt.show()

# Display the merged and sorted DataFrame
print("Merged Department Budget and Expenses with Variance:\n")
print(df_sorted)

# Plot (Optional)
plot_budget_vs_actual(df_sorted)
```
