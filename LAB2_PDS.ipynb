```python
import numpy as np
import matplotlib.pyplot as plt

months = np.array([
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
])

temps_celsius = np.array([12.1, 14.3, 18.5, 22.0, 26.4, 30.1,
                          31.2, 30.5, 28.0, 23.1, 17.0, 13.5])

average_temp = np.mean(temps_celsius)
max_temp = np.max(temps_celsius)
min_temp = np.min(temps_celsius)
hottest_month = months[np.argmax(temps_celsius)]
coldest_month = months[np.argmin(temps_celsius)]

normalized_temps = (temps_celsius - min_temp) / (max_temp - min_temp)
temps_fahrenheit = temps_celsius * 9 / 5 + 32
above_average = months[temps_celsius > average_temp]

print("Monthly Temperatures (°C):", temps_celsius)
print(f"\nAverage Temperature: {average_temp:.2f}°C")
print(f"Hottest Month: {hottest_month} ({max_temp:.1f}°C)")
print(f"Coldest Month: {coldest_month} ({min_temp:.1f}°C)")
print("\nNormalized Temperatures:", normalized_temps.round(2))
print("\nTemperatures in Fahrenheit:", temps_fahrenheit.round(1))
print("\nMonths Above Average Temperature:", above_average)

plt.plot(months, temps_celsius, marker='o', label="°C")
plt.plot(months, temps_fahrenheit, marker='s', linestyle='--', label="°F")
plt.title("Monthly Temperature Trend")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

data = {
    'Order ID': [101, 102, 103, 104, 105, 106],
    'Product': ['Laptop', 'Mobile', 'Tablet', 'Laptop', 'Headphones', 'Mobile'],
    'Quantity': [2, 5, 3, 1, 10, 4],
    'Price': [75000, 15000, 20000, 80000, 1500, 16000]
}

df = pd.DataFrame(data)

df['Total Sale'] = df['Quantity'] * df['Price']

high_value_orders = df[df['Total Sale'] > 1000]

product_sales = df.groupby('Product')['Quantity'].sum().reset_index()
most_sold_product = product_sales.sort_values(by='Quantity', ascending=False).iloc[0]

def calculate_discount(sale):
    if sale > 100000:
        return 0.10 * sale
    elif sale > 50000:
        return 0.05 * sale
    else:
        return 0

df['Discount'] = df['Total Sale'].apply(calculate_discount)

df.to_excel("ecommerce_sales_summary.xlsx", index=False)

print("Sales Data:\n", df)
print("\nHigh Value Orders:\n", high_value_orders)
print(f"\nMost Sold Product: {most_sold_product['Product']} ({most_sold_product['Quantity']} units)")
```
